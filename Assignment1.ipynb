```python
# Task 6 - Adding HDI Category Column
# Function to classify countries based on UNDP HDI thresholds
def classify_hdi(hdi_value):
    if hdi_value < 0.550:
        return "Low"
    elif 0.550 <= hdi_value <= 0.699:
        return "Medium"
    elif 0.700 <= hdi_value <= 0.799:
        return "High"
    else:
        return "Very High"

# Creating a new column 'HDI Category' based on HDI values
hdi_2022_df['HDI Category'] = hdi_2022_df['hdi'].apply(classify_hdi)

# Checking the distribution of HDI categories
hdi_2022_df['HDI Category'].value_counts()

# Displaying sample rows to verify classification
hdi_2022_df[['country', 'hdi', 'HDI Category']].head(10)
```

```python
# Task - 2. Composite Development Score

hdi_sa = pd.read_csv("/content/drive/MyDrive/Assignment - 1/HDI_SouthAsia.csv")

# Createing Composite Score
# Formula: Composite Score = 0.3 * Life Expectancy Index + 0.3 * GNI per Capita Index
hdi_sa['composite_score'] = 0.3 * hdi_sa['life_expectancy'] + 0.3 * hdi_sa['gross_inc_percap']


# Rank countries by Composite Score (descending)
hdi_sa['composite_rank'] = hdi_sa['composite_score'].rank(ascending=False, method='min')

# Sort the dataset by Composite Score to show ranking
ranking_table = hdi_sa[['country', 'composite_score', 'composite_rank']].sort_values(by='composite_score', ascending=False)

# Reset index for neat display
ranking_table = ranking_table.reset_index(drop=True)

#printing ranking
print("South Asian Countries Ranked by Composite Score:")
print(ranking_table)
```

```python
# Ploting the top 5 countries
import pandas as pd
import matplotlib.pyplot as plt

# Aggregate by country (take the latest year for each country)
latest_data = hdi_sa.sort_values('year').groupby('country').tail(1)

# Rank countries by Composite Score
top5 = latest_data.sort_values(by='composite_score', ascending=False).head(5)

# Plot horizontal bar chart
plt.figure(figsize=(10,6))
plt.barh(top5['country'], top5['composite_score'], color='skyblue')
plt.xlabel('Composite Score')
plt.title('Top 5 South Asian Countries by Composite Score')
plt.gca().invert_yaxis()  # Highest score on top
plt.show()

# print top 5 table
print("Top 5 South Asian countries by Composite Score:")
print(top5[['country', 'year', 'composite_score']])
```

```python
# Compareing the ranking of countries by Composite Score with their HDI ranking and discuss any differences.
# Create a new column to show difference
import pandas as pd

# Assumeing top5 comes from latest_data after calculating composite_score
# Ensure HDI rank exists
top5['hdi_rank'] = top5['hdi'].rank(ascending=False, method='min')

# Create rank difference
top5['rank_difference'] = top5['hdi_rank'] - top5['composite_rank']

# Print comparison
print("Top 5 Countries: Rank Differences (HDI Rank - Composite Rank)")
print(top5[['country', 'composite_rank', 'hdi_rank', 'rank_difference']])
```

```python
# Task 3. Outlier Detection

# Detect outliers using 1.5 × IQR rule(Interquartile Range rule)

def detect_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Return boolean series where True indicates an outlier
    return (series < lower_bound) | (series > upper_bound)

# Detect outliers
hdi_sa['hdi_outlier'] = detect_outliers(hdi_sa['hdi'])
hdi_sa['gni_outlier'] = detect_outliers(hdi_sa['gross_inc_percap'])
```

```python
# Scatter plot of GNI per Capita vs HDI
plt.figure(figsize=(10,6))
plt.scatter(hdi_sa['gross_inc_percap'], hdi_sa['hdi'], color='blue', label='Normal')
plt.scatter(hdi_sa.loc[hdi_sa['hdi_outlier'] | hdi_sa['gni_outlier'], 'gross_inc_percap'],
            hdi_sa.loc[hdi_sa['hdi_outlier'] | hdi_sa['gni_outlier'], 'hdi'],
            color='red', label='Outlier')
plt.xlabel('GNI per Capita')
plt.ylabel('HDI')
plt.title('Scatter Plot of GNI per Capita vs HDI with Outliers Highlighted')
plt.legend()
plt.show()

# Print outlier countries
outliers = hdi_sa[hdi_sa['hdi_outlier'] | hdi_sa['gni_outlier']]
print("Outlier Countries (HDI or GNI per Capita):")
print(outliers[['country', 'year', 'hdi', 'gross_inc_percap', 'hdi_outlier', 'gni_outlier']])
```

```python
# Task 4: Exploring Metric Relationships

# Select metrics
metrics = ['life_expectancy', 'gender_development']
hdi_column = 'hdi'


# Compute Pearson correlation with HDI
print("Pearson Correlation with HDI:")
for metric in metrics:
    corr = hdi_sa[metric].corr(hdi_sa[hdi_column])
    print(f"{metric}: {corr:.3f}")
```

```python
# Scatter plots with trendlines using seaborn
import matplotlib.pyplot as plt
import seaborn as sns

for metric in metrics:
    plt.figure(figsize=(8,6))
    sns.regplot(x=metric, y=hdi_column, data=hdi_sa,
                scatter_kws={'color':'blue'}, line_kws={'color':'red'})
    plt.xlabel(metric)
    plt.ylabel('HDI')
    plt.title(f'{metric} vs HDI')
    plt.show()
```

Discussion:
Life Expectancy Index is most strongly related to HDI, as countries with higher life expectancy generally have higher HDI values. In contrast, Gender Development Index shows the weakest relationship, indicating that gender equality contributes less directly to HDI in the region. This suggests that health and income are the main drivers of human development, while gender development has a smaller impact on overall HDI rankings.

```python
# Ploting top 3 positive gaps
plt.figure(figsize=(8,5))
plt.bar(gap_desc['country'].head(3), gap_desc['GNI_HDI_Gap'].head(3), color='green')
plt.ylabel('GNI HDI Gap')
plt.title('Top 3 Positive GNI HDI Gaps (GNI > HDI)')
plt.show()

# Ploting top 3 negative gaps
plt.figure(figsize=(8,5))
plt.bar(gap_asc['country'].head(3), gap_asc['GNI_HDI_Gap'].head(3), color='red')
plt.ylabel('GNI HDI Gap')
plt.title('Top 3 Negative GNI HDI Gaps (GNI < HDI)')
plt.show()
```

Discussion Points

May suggest that high income is not translating into better health, education, or quality of life.

Negative Gap:

Shows efficiency in converting income into human development outcomes.

```python
# Task - 7. Outlier Detection
# Function to find outliers using IQR rule
def find_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    return df[(df[column] < lower) | (df[column] > upper)]

# Outliers
sa_hdi_outliers = find_outliers(hdi_south_asia, 'hdi')
me_hdi_outliers = find_outliers(hdi_middle_east, 'hdi')

# Scatter plot with outliers highlighted
plt.figure(figsize=(10,6))
plt.scatter(hdi_south_asia['gross_inc_percap'], hdi_south_asia['hdi'], label='South Asia', color='blue')
plt.scatter(sa_hdi_outliers['gross_inc_percap'], sa_hdi_outliers['hdi'], color='red', label='SA Outliers')
plt.scatter(hdi_middle_east['gross_inc_percap'], hdi_middle_east['hdi'], label='Middle East', color='green')
plt.scatter(me_hdi_outliers['gross_inc_percap'], me_hdi_outliers['hdi'], color='orange', label='ME Outliers')
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("HDI vs GNI per Capita with Outliers Highlighted")
plt.legend()
plt.show()
```

Discussion:

Highlight countries with unusually high or low HDI relative to GNI.

Possible causes: economic inequality, health crises, education gaps.
